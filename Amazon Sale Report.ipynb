# Highest Selling Product Category

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=10,edgecolor='Black')
plt.ylabel('Count')
plt.xlabel('Category')
# plt.xticks(rotation=90)
plt.show()
```

# Checking B2B data

```python
B2B_data = df['B2B'].value_counts()
plt.pie(B2B_data, labels = B2B_data.index, autopct = '%1.1f%%',wedgeprops={'edgecolor': 'black'})
plt.show()
```

**Note: From above chart one thing is clear that maximum 99.3% of buyers are retailers and 0.7% are B2B buyers**

# Size Availability

```python
x_axis = df['Category']
y_axis = df['Size']
plt.figure(figsize=(10,5))
plt.scatter(x_axis,y_axis)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Size Availability')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 📌 Observation from "Size Availability Across Categories" Graph

- The scatter plot shows the availability of different **product sizes** across various **categories**.
- Some categories (like **T-shirt**, **Shirt**) offer a **wide range of sizes**, while others (like **Shoes** or **Watch**) may have only **one or two sizes**.
- Categories with **limited or no dots** (e.g., rare size options) may have **less variety** or focus on **standard sizing**.

# Distribution of states

```python
first_ten_states = df['ship-state'].value_counts().head(10).index

plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(first_ten_states)], x='ship-state', hue='ship-state',order=first_ten_states,palette='Set2')

plt.ylabel('Count')
plt.title('Top 10 Shipping States')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Note: From the above graph, we can see that most of the buyers are from the Maharashtra state.**
